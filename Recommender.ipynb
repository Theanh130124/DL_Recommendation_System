```python
# id,title,genre,overview  -> day la cac cot quan trong , cac cot con lai loai bo
movies = movies[['id','title','genre','overview']]
movies
```

Feature Section Part

```python
movies['tags']=movies['overview']+movies['genre'] #nhom chung thanh tag
movies
```

```python
new_data = movies.drop(columns=['overview','genre']) #Giu id va title de truyen qua API de lay ra ds phim (vd tra ve ten phim )
new_data
```

```python
#NLP
# 1.Bag of word
# 2.TFIDE
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000,stop_words='english') #Chuyen doi vanban thanh vector cho may hieu  # loại bỏ 'this', 'is', 'in', 'and' do english
vector = cv.fit_transform(new_data['tags'].values.astype('U')).toarray() # values.astype('U'): ép kiểu dữ liệu về unicode string (bắt buộc nếu có NaN hoặc dữ liệu không chuẩn).

#fit_transform(): học các từ xuất hiện trong cột tags, rồi chuyển thành vector.


print(cv.get_feature_names_out()) #IN ra bag of word no lay duoc
print(vector)   #in ra mang vector
```
