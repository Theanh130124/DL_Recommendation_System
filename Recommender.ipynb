```python
# id,title,genre,overview  -> day la cac cot quan trong , cac cot con lai loai bo
movies = movies[['id','title','genre','overview']]
movies
```

Feature Section Part

```python
movies['tags']=movies['overview']+movies['genre'] #nhom chung thanh tag
movies
```

```python
new_data = movies.drop(columns=['overview','genre']) #Giu id va title de truyen qua API de lay ra ds phim (vd tra ve ten phim )
new_data
```

```python
#NLP
# 1.Bag of word
# 2.TFIDE
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000,stop_words='english') #Chuyen doi vanban thanh vector cho may hieu  # loại bỏ 'this', 'is', 'in', 'and' do english
vector = cv.fit_transform(new_data['tags'].values.astype('U')).toarray() # values.astype('U'): ép kiểu dữ liệu về unicode string (bắt buộc nếu có NaN hoặc dữ liệu không chuẩn).

#fit_transform(): học các từ xuất hiện trong cột tags, rồi chuyển thành vector.


print(cv.get_feature_names_out()) #IN ra bag of word no lay duoc
print(vector)   #in ra mang vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector) #Tinh do tuong dong cosin
similarity
```

```python
# #Code nay test dung chay lai

# #Lay chi muc theo title
# new_data[new_data['title'] =="The Godfather"].index[0]
# # new_data[new_data['title'] =="The Godfather"].index[0] #lay id
```

 #Lấy danh sách tất cả phim và sắp xếp chúng theo độ tương đồng với phim thứ 2 (index = 2) trong ma trận similarity.

#Dòng này giúp chọn ra những phim nào giống với phim có index = 2 nhất (ở đây bạn nên thay bằng index của "The Godfather" để đúng mục tiêu).

```python
#Code nay test dung chay lai

distance = sorted(list(enumerate(similarity[2])),reverse=True , key=lambda vector:vector[1]) #keys=lambda vector:vector[1] do similarities = [(0, 0.9), (1, 0.3), (2, 0.8)] la index , score
#lay vector[1] chinh la lay score de so sanh ,,
#tao ranh enumerate(similarity[2] tuple(index,score)

#Truy cap 5 phim gan nhat (giong nhat)
for i in distance[1:6]: # bỏ i=0 vì đó chính là phim đang chọ
    print(new_data.iloc[i[0]].title) #Lay hang thu 0 cua cot title  vd : df.iloc[2]['year']   # Lấy năm của dòng thứ 3: 2010

```

```python
#Fuction recommender
def recommand(movies):
  index = new_data[new_data['title'] == movies].index[0]
  distance = sorted(list(enumerate(similarity[index])),reverse=True , key=lambda vector:vector[1]) #similarity index
  for i in distance[1:6]:
    print(new_data.iloc[i[0]].title)
```

```python
recommand("Iron Man")
```
