```python
import numpy as np
import pandas as pd
from marshmallow import missing

books = pd.read_csv('books_cleaned.csv')

```

```python
#Phan loai theo  Fiction va NonFiction (Hu cau voi phi hu cau)

category_mapping = {'Fiction':"Fiction",
                    'Juvenile Fiction':"Children's Fiction",
                    'Biography & Autobiography':"Nonfiction",
                    'History':"Nonfiction",
                    'Literary Criticism	':"Nonfiction",
                    'Religion':"Nonfiction",
                    'Philosophy':"Nonfiction",
                    'Comics & Graphic Novels':"Fiction",
                    'Drama':"Fiction",
                    'Juvenile Nonfiction':"Nonfiction",
                    'Science':"Nonfiction",
                    'Poetry':"Fiction",}

books['simple_categories'] = books['categories'].map(category_mapping)
books
```

```python
books[~(books["simple_categories"].isna())] #Bo null  nhung thang co simple_categories di
```

```python
from transformers import pipeline

fiction_categories = ["Fiction","Nonfiction"]

pipe = pipeline("zero-shot-classification" ,
                model="facebook/bart-large-mnli",
                device="cpu",)
```

```python
sequence = books.loc[books["simple_categories"] == "Fiction","description"].reset_index(drop=True)  #Lọc hàng (rows) trong DataFrame books nơi cột "simple_categories" có giá trị "Fiction".

# Sau đó lấy cột "description" tương ứng (tức là mô tả của các sách thuộc thể loại "Fiction").
```

```python
pipe(sequence,fiction_categories) #-> tinh xac xuat nhan
# Gọi pipeline zero-shot-classification đã được tạo ở trước (pipe) để phân loại danh sách mô tả (sequence) vào các nhãn "Fiction" và "Nonfiction".
```

```python
import numpy as np
max_index = np.argmax(pipe(sequence,fiction_categories)["scores"])
max_label = pipe(sequence,fiction_categories)["labels"][max_index]
max_label

#Luu chi so voi nhan cao nhat lai
```

```python
def generate_predictions(sequence,categories):
    predictions = pipe(sequence,categories) #Gọi mô hình
    max_index = np.argmax(predictions["scores"]) #Lấy score cao nhất
    max_label = predictions["labels"][max_index] #Lấy tên nhãn
    return max_label

# được dùng để dự đoán nhãn (label) có xác suất cao nhất từ một chuỗi văn bản (sequence) trong số các nhãn cho trước (categories
```

```python
from tqdm import tqdm  #Thanh bar tiến trình
#
# tqdm(range(0, 300)): tạo thanh tiến trình cho vòng lặp từ 0 đến 299.
#
# books.loc[books["simple_categories"] == "Fiction", "description"]: lọc ra các dòng có thể loại đơn giản là "Fiction", lấy cột description.
#
# .reset_index(drop=True)[i]: đặt lại chỉ số để truy cập đúng dòng i từ 0 → 299.
#
# generate_predictions(sequence, fiction_categories): gọi hàm đã định nghĩa trước đó để dự đoán nhãn từ mô hình (với danh sách fiction_categories — danh sách các nhãn fiction có thể).
#
# prediction_cats += [...]: thêm nhãn được dự đoán vào danh sách kết quả.
#
# actual_cats += ["Fiction"]: thêm nhãn thực tế để sau này đánh giá độ chính xác.

actual_cats =[] #Danh mục thực te
prediction_cats = [] #Danh mục dô mô hình dự đoán

for i in tqdm(range(0,300)):
    sequence = books.loc[books["simple_categories"] == "Fiction","description"].reset_index(drop=True)[i]
    prediction_cats += [generate_predictions(sequence,fiction_categories)]
    actual_cats += ["Fiction"]
```

```python
for i in tqdm(range(0,300)):
        sequence = books.loc[books["simple_categories"] == "Nonfiction","description"].reset_index(drop=True)[i]
        prediction_cats += [generate_predictions(sequence,fiction_categories)]
        actual_cats += ["Nonfiction"]
```

```python
#Lưu các ket qua

prediction_df = pd.DataFrame({"actual_categories": actual_cats,"predicted_categories":prediction_cats})
prediction_df
```

```python
prediction_df["correct_prediction"] = (np.where(prediction_df["actual_categories"] == prediction_df["predicted_categories"],1,0)) #In ra 1 0 nhìn cho dễ
```

```python
prediction_df["correct_prediction"].sum()/len(prediction_df) #đúng 82%
```

```python
isbns = []
prediction_cats = []
missing_cats = books.loc[books["simple_categories"].isna() , ["isbn13","description"]].reset_index(drop=True)


# Tìm tất cả các dòng trong DataFrame books mà simple_categories bị thiếu (NaN).
#
# Giữ lại chỉ 2 cột: "isbn13" và "description" để sử dụng cho việc dự đoán.
#
# reset_index(drop=True) để đặt lại chỉ số dòng từ 0, tránh lỗi về index sau này.
```

```python
for i in tqdm(range(0,len(missing_cats))):
    sequence = missing_cats["description"][i]
    prediction_cats += [generate_predictions(sequence,fiction_categories)]
    isbns += [missing_cats["isbn13"][i]]
```

```python
missing_predicted_df = pd.DataFrame({"isbn13":isbns,"predicted_categories":prediction_cats})
```

```python
books = pd.merge(books,missing_predicted_df,how="left",on="isbn13")
books["simple_categories"] = np.where(books["simple_categories"].isna(),books["predicted_categories"],books["simple_categories"])
books = books.drop(columns="predicted_categories")
```
